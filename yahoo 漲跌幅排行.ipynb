# 抓取 Yahoo 的漲跌幅排行(上市)

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

url = 'https://tw.stock.yahoo.com/rank/change-up?exchange=TAI'
headers = {'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'}
response = requests.get(url, headers = headers)
if response.status_code == 200:
    text =  response.text
else:
    print('錯誤代碼',response.status_code)
    
data = pd.DataFrame({'名稱':[],'排行':[],'股票':[],'成交價':[],'漲跌':[],'漲跌幅':[],'最高':[],'最低':[],'價差':[],'成交量':[],'成交值(億)':[]})

soup = BeautifulSoup(text, 'html.parser')
all_stock = soup.find_all('li',class_ = 'List(n)')
for stock in all_stock:
    this_stock = []
    stock_name = stock.find('div',class_='Ell').text
    this_stock.append(stock_name)
    stock_detail = stock.find_all('span')
    for i in stock_detail:
        if i.text != '':
            this_stock.append(i.text)
    data.loc[len(data)] = this_stock
```
